```python
combine=pd.merge(df,all_sites,on='FILM',how='inner')
```

```python
combine.head()
```

```python
import numpy as np
combine['Rot_Norm']=np.round(combine['RottenTomatoes']/20,1)
combine['RTU_Norm']=np.round(combine['RottenTomatoes_User']/20,1)
```

```python
import numpy as np
combine['Meta_Norm']=np.round(combine['Metacritic']/20,1)
combine['Meta_U_Norm']=np.round(combine['Metacritic_User']/2,1)
```

```python
combine['IMDB_Norm']=np.round(combine['IMDB']/2,1)
```

```python
combine.head()
```

```python
norm_scores=combine[['STARS','RATING','Rot_Norm','RTU_Norm','Meta_Norm','Meta_U_Norm','IMDB_Norm']]
```

```python
norm_scores.head()
```

```python
plt.figure(figsize=(15,6),dpi=200)
sns.kdeplot(data=norm_scores,clip=[0,5],shade=True,palette='Set1')
```

```python
fig,ax=plt.subplots(figsize=(15,6),dpi=200)
sns.kdeplot(data=norm_scores[['Rot_Norm','STARS']],clip=[0,5],shade=True,palette='Set1')
```

```python
sns.histplot(norm_scores,bins=50)
```

```python
sns.clustermap(norm_scores,cmap='magma')
```
